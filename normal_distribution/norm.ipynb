Провести многократные измерения заданного интервала времени - 5 секунд.

```python
%matplotlib inline  

import pandas

import matplotlib
import numpy
import matplotlib.pyplot as plt

df = pandas.read_csv("measuring.csv")
df['measuring'].plot(kind='bar')
print(df.measuring.describe())
df.plot.kde()
from scipy import stats
print(stats.kstest(df, 'norm', (df.mean(), df.std()), N=len(df)))
```
